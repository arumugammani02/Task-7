```python
for k, v in df1.items():
  q1 = v.quantile(0.25)
  q2 = v.quantile(0.75)
  irq = q2 - q1
  v_col = v[(v <= q1 - 1.5 * irq) | (v >= q2 + 1.5 * irq)]
  per = np.shape(v_col)[0] * 100.0 / np.shape(df1)[0]
  print("Column %s outliers = %.4f%%" % (k, per))
```

```python
def outlierpresence(df1):
  for i in df1.keys():
    Q1 = df1.quantile(0.25)
    Q2 = df1.quantile(0.75)
    IQR = Q2 - Q1
    x = (df1 < (Q1 - 1.5 * IQR)) |(df1 > (Q2 + 1.5 * IQR))
    substring = 'True'

# It will give all outliers in the dataframe with all columns    
    y= x[x.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]

#Here we check true is in the list of particular column
    if True in y[i].tolist():
      print('Outliers', '\033[1m'+ 'present' +'\033[0m', 'in the data of','\033[1m' + i + '\033[0m')
      print('-------------------------------')
    else:
      print('Outliers', '\033[1m'+ ' not present in the data of' +'\033[0m', 'in','\033[1m' + i + '\033[0m') 
      print('-------------------------------') 
outlierpresence(df1)
```


## **Euclidean distance metrics**

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
y_train = df1['Type_Of_Glass']
X_train = df1.iloc[:,0:9]
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=6,metric = 'euclidean')
knn.fit(X_train, y_train) # train the model
print("Train Accuracy : ", knn.score(X_train,y_train)*100)
print("Value Accuracy : ", np.mean(cross_val_score(knn, X_train, y_train, cv=6)))
```
